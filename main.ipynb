## 变换矩阵

```python
# 变换矩阵T X轴放大2倍，Y轴放大1.5倍
T = Matrix([[2, 0], [0, 1.5]])
def show(T):
    import matplotlib.pyplot as plt
    points = [[0, 0], [0, 5], [3, 5], [3, 4], [1, 4], [1, 3], [2, 3], [2, 2], [1, 2], [1, 0]]
    x = [point[0] for point in points]
    y = [point[1] for point in points]
    plt.figure(figsize=(5,5))
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.plot(x, y)
    P = Matrix(points)
    # 先转置P为2*10矩阵，再被T转换成P2
    P2 = T.dot(P.T())
    plt.plot([P2.col_vector(i)[0] for i in range(P2.col_num())],
             [P2.col_vector(j)[1] for j in range(P2.col_num())])
    plt.show()
    
show(T)
```

```python
# 关于X轴翻转（X不变，Y轴相反）
T = Matrix([[1, 0], [0, -1]])
show(T)
```

```python
# 关于Y轴翻转（X变反，Y轴不变）
T = Matrix([[-1, 0], [0, 1]])
show(T)
```

```python
# 关于原点翻转（X不变，Y轴相反）
T = Matrix([[-1, 0], [0, -1]])
show(T)
```

### 错切

```python
# 沿着X轴错切
T = Matrix([[1, 0.5], [0, 1]])
show(T)
```

```python
# 沿着Y轴错切
T = Matrix([[1, 0], [0.5, 1]])
show(T)
```

```python
import math
theta = math.pi / 3
# 沿着原点顺时针旋转60度
T = Matrix([[math.cos(theta), math.sin(theta)],[-math.sin(theta),math.cos(theta)]])
show(T)
```

## 单位矩阵

```python
# 自定义创建单位矩阵
I= Matrix.identity(3)
B = Matrix([[3,7,1],[8,4,7],[5,6,3]])
print('I:{}'.format(I))
print('B:{}'.format(B))
print('I.dot(B):{}'.format(I.dot(B)))
print('B.dot(I):{}'.format(B.dot(I)))
```

```python
import numpy as np
I = np.identity(2)
A = np.array([[10, 100],[200, 300]])

print('I:{}'.format(I))
print('A:{}'.format(A))
print('I.dot(A):{}'.format(I.dot(A)))
print('A.dot(I):{}'.format(A.dot(I)))
```

## 逆矩阵

```python
# 只有方针才可能是可逆的
import numpy as np
A = np.array([[1, 2],[3, 4]])
invA = np.linalg.inv(A)
print('invA:{}'.format(invA))
# 在计算浮点计算里有精度差异
print('invA.dot(A):{}'.format(invA.dot(A)))
print('A.dot(invA):{}'.format(A.dot(invA)))
```
